# Webscrape on eBay

* The goal for this notebook is to showcase how to webscrape listing items on eBay using BeautifulSoup. We search for Jeep data

### 1. Get webpages using requests

```python
import requests
req = requests.get("https://www.ebay.com/sch/i.html?_from=R40&_trksid=p4432023.m570.l1313&_nkw=jeep&_sacat=0")
```

```python
webpage = req.text
```

### 2.Get specific contents using BeautifulSoup

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(webpage, 'html.parser')
```

### 2.1 Prettify the webpage

```python
print(soup.prettify)
```

### 2.2 Get the paragraphs

```python
listing_items = soup.find_all("div", class_="s-item__info clearfix")
```

```python
listing_items
```

```python
data = {"title":[], "url":[], "price":[]}
for item in listing_items:
    # find the texts
    title = item.find("div", class_="s-item__title").text.strip()
    url = item.find("a", class_="s-item__link")["href"]
    price = item.find("span", class_="s-item__price").text.strip()
    
    # append to the lists inside the dictonary
    data["title"].append(title)
    data["url"].append(url)
    data["price"].append(price)
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
df
```

```python
df.to_csv('Sample_data.csv', index=False)
```
